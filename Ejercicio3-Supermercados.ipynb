```
#funciones auxiliares
def obtener_fecha(nombre_fichero):
    valores = nombre_fichero.split("_")
    # (TODO): a partir de la lista valores, identificar el día y el mes y pasarlos a int
    #supermercado_1_02092020.txt
    print (valores)
    dia = int(valores[2][0:2])
    mes = int(valores[2][2:4])
    print ('mes:', mes, 'dia:', dia)
    return dia, mes

def leer_fichero(ruta_fichero):
    info = []
    with open(ruta_fichero, 'r') as f:
        for line in f:
            info.append(line.strip())
    return info

def posicion_columna(campo):
    # mapeo identifica el campo con su posición en el fichero
    mapeo = {'Supermarket': 1, 'day': 2, 'month': 3, 'Fresh': 4, 'Milk': 5, 'Grocery': 6,
             'Frozen': 7, 'Detergents_paper': 8, 'Delicatessen': 9}
    return mapeo[campo]
```

```
#add_new_line.py
import openpyxl as opx
#from config import ruta_fichero_diario, ruta_maestro, nombre_hoja, supermercado_importar
#from funciones_auxiliares import obtener_fecha, leer_fichero, posicion_columna

class ImportarDatos:

    def __init__(self, ruta_fichero, reporte, hoja):
        self.ruta_fichero = ruta_fichero
        self.reporte = reporte
        self.hoja = hoja
        self.datos = {}

    def leer_supermercado1(self):
        # leer el fichero
        info = leer_fichero(self.ruta_fichero)

        # crear diccionario para guardar los datos
        datos = {}

        # leer linea a linea
        for dato in info:

            # (TODO): usar la funcion de .split() para obtener el nombre del campo y el valor asociado
            campo, valor = dato.split(' ')

            # (TODO): campo convertirlo a Capitalize (por ejemplo, FRESH => Fresh) y el valor a int:
            campo = campo.capitalize()
            valor = int(valor)

            datos[campo] = valor

        # añadir el campo Supermercado y su valor
        datos['Supermarket'] = 1

        return datos

    def leer_supermercado2(self):
        # (TODO): completar esta función
        # abrir fichero:
        info = leer_fichero(self.ruta_fichero)
        texto = info[0]
        texto = texto.split(',')

        # crear diccionario para guardar los datos
        datos = {}

        # leer linea a linea
        for dato in texto:

            # (TODO): usar la funcion de .split() para obtener el nombre del campo y el valor asociado
            campo, valor = dato.split(':')

            # (TODO): campo convertirlo a Capitalize (por ejemplo, FRESH => Fresh) y el valor a int:
            campo = campo.capitalize()
            valor = int(valor)

            datos[campo] = valor

        datos['Supermarket'] = 2

        return datos

    # leer datos diarios:
    def leer_fichero_diario(self, supermercado):
        if supermercado == 1:
            datos = self.leer_supermercado1()
        else:
            datos = self.leer_supermercado2()

        dia, mes = obtener_fecha(self.ruta_fichero)

        datos['day'] = dia
        datos['month'] = mes

        self.datos = datos
        return

    def insertar_datos_diarios(self):
        # (TODO): cargar el workbook y la hoja donde vamos a insertar los datos
        workbook = opx.load_workbook(ruta_maestro) 
        sheet = workbook[nombre_hoja]

        # max_row identifica la última linea
        newRowLocation = sheet.max_row + 1

        # añade la información requerida:
        for campo in self.datos.keys():
            # (TODO): incluir los datos correspondientes en cada una de las celdas, 
            # utilizar la función posicion_columna 
            # para identificar la columna a partir del campo
            row1 = newRowLocation
            col1 = posicion_columna(campo)
            val1 = self.datos[campo]
            print (row1, col1, val1)

            sheet.cell(row = newRowLocation, column = posicion_columna(campo), value = self.datos[campo])

        workbook.save(filename=self.reporte)
        workbook.close()
        return


importar = ImportarDatos(ruta_fichero_diario, ruta_maestro, nombre_hoja)

importar.leer_fichero_diario(supermercado_importar)
importar.insertar_datos_diarios()
```
